```python
import numpy as np 
import pandas as pd 
import ast
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Load datasets
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
# Display initial data
movies.head(2)

```

```python
# Merge datasets on 'title'
movies = movies.merge(credits, on='title')
movies.head()
```

```python
# Select relevant columns
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew', 'vote_average']]
```

```python
# Function to convert JSON strings to lists of names
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
# Remove rows with missing values
movies.dropna(inplace=True)
```

```python
# Apply the conversion function to genres and keywords columns
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
# Function to convert the first 3 names in the cast
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter += 1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
# Function to fetch director's name
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python

ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python

movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
# Function to remove spaces in names
def collapse(L):
    return [i.replace(" ", "") for i in L]
```

```python
# Apply the collapse function to the columns
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
# Split the overview into words
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
# Create a new 'tags' column by combining all relevant information
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
# Create a new dataframe with only the necessary columns
new = movies.drop(columns=['overview', 'keywords', 'cast', 'crew'])
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
```

```python
# Include the 'genres' column in the new dataframe
new['genres'] = movies['genres']  # Add this line
```

```python
# Vectorize the tags
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
vector = cv.fit_transform(new['tags']).toarray()
```

```python
# Calculate cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
# Recommendation function based on movie title or genre
def recommend(movie=None, genre=None):
    if genre:
        # Filter movies by genre
        genre_movies = new[new['genres'].apply(lambda x: genre in x)]
        if genre_movies.empty:
            print(f"No movies found for genre: {genre}")
            return
        
        # Sort genre movies by vote_average
        sorted_genre_movies = genre_movies.sort_values(by='vote_average', ascending=False)
        
        # Recommend the top 10 rated movies in the selected genre
        print(f"Top 10 rated movies in the genre '{genre}':")
        for idx, row in sorted_genre_movies.head(10).iterrows():
            print(f"{row['title']} - Rating: {row['vote_average']}/10")
    
    elif movie:
        if movie not in new['title'].values:
            print(f"Movie '{movie}' not found in the dataset.")
            return
        
        # Find the index of the movie
        index = new[new['title'] == movie].index[0]
        distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
        
        # Print the top 5 similar movies
        print(f"Top 5 recommendations similar to '{movie}':")
        for i in distances[1:6]:  # Skip the first movie because it is the same movie
            movie_title = new.iloc[i[0]].title
            movie_rating = new.iloc[i[0]].vote_average
            print(f"{movie_title} - Rating: {movie_rating}/10")
```

```python
# Example usage
recommend(movie="Iron Man 2")
```

```python
recommend(genre="ScienceFiction")
```
